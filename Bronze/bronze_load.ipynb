```python
from pyspark.sql import SparkSession

df_airlines=spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/workspace/default/flights/airlines.csv")
df_airlines.write.format("delta").mode("overwrite").saveAsTable("flight_bronze.bronze_airlines")
spark.sql("SELECT * FROM flight_bronze.bronze_airlines LIMIT 10").show()
```
